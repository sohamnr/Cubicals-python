DATA VISUALIZATION 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Scatter plot

```python
x_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100

#print(x_data)
#print(y_data)

plt.scatter(x_data, y_data, s=80, marker="*", alpha = 0.3) 
plt.show()

```

Line plot

```python
years = [2004 + x for x in range(22)]
weight = [78,78,79,81,81,83,83,85,86,89,92,95,88,86,85,82,82,81,80,80,79,78]

plt.plot(years, weight, c="green", lw=5)
plt.xlabel("years")
plt.ylabel("weights(in kgs)")
plt.show()
```

bar plot

```python
x = ["C", "c++", "Pyhton", "Java", "Javascript", "HTML", "CSS"]
y = [10,15,70,55,67,40,33]
plt.bar(x, y, width = 0.5, color = "red",  edgecolor = "black", lw = 3)
plt.show()
```

Histogram plot

```python
ages = np.random.normal(25, 1.2, 500)
print(ages)
plt.hist(ages, cumulative= True)
```

pie chat plot

```python
langs = ["C", "c++", "Pyhton", "Java", "Javascript", "HTML", "CSS"]
votes = [10,15,70,55,67,40,33]
explodes = [0, 0.3, 0.3, 0, 0,0, 0 ]
plt.pie(votes, labels=langs, explode=explodes, autopct="%.2f%%")
```

box plot

```python
heights = np.random.normal(172, 8, 300)
print(heights)
plt.boxplot(heights)
```

multiple plots

```python
stock_a = [110, 110, 112, 113, 109, 109, 112]
stock_b = [100, 104, 103, 106, 109, 107, 102]
stock_c = [112, 117, 118, 120, 115, 108, 116]

plt.plot(stock_a, label="company1")
plt.plot(stock_b, label="company2")
plt.plot(stock_c, label="company3")
plt.legend()

```

3D plotting

```python
ax = plt.axes(projection='3d')

x1 = np.random.rand(100)  
y1 = np.random.rand(100)  
z1 = np.random.rand(100)  
ax.scatter(x1, y1, z1)

ax.set_title("3D plotting")
plt.show()


```
